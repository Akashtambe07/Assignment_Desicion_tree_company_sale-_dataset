```python
import pandas
import numpy
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets  
from sklearn.model_selection import train_test_split
```

```python
company_sale=pandas.read_csv('Company_Data.csv')
company_sale
```

```python
company_sale.info()
```

```python
company_sale.describe()
```

```python
company_sale.isnull().sum()
```

```python
sns.countplot(company_sale.ShelveLoc)
```

```python
# Converting Target variable 'Sales' into categories Low, Medium and High.
company_sale['Sales'] = pandas.cut(company_sale.Sales,bins=[0,6,12,17], labels=['Low','Medium', 'High'], right = False)
company_sale
```

```python
company_sale.Sales.value_counts()
```

```python
company_sale
```

```python
#converting categorical data into numeric 
from sklearn.preprocessing import LabelEncoder
Encoder=LabelEncoder()
```

```python
company_sale['Sales']=Encoder.fit_transform(company_sale["Sales"])
company_sale['ShelveLoc']=Encoder.fit_transform(company_sale["ShelveLoc"])
company_sale['Urban']=Encoder.fit_transform(company_sale["Urban"])
company_sale['US']=Encoder.fit_transform(company_sale["US"])
```

```python
company_sale
```

```python
company_sale['Sales'].unique()
```

```python
#dividing data into x & y variable 
x=company_sale.iloc[:,1:11]

y=company_sale['Sales']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=23)
```

# Building Decision Tree Classifier using Entropy Criteria[C5.0]

```python
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
```

```python
model=DecisionTreeClassifier(criterion="entropy",max_depth=3)
model.fit(x_train,y_train)
```

```python
#PLot the decision tree
plt.figure(figsize=(18,10))

tree.plot_tree(model);
```

```python
fig, axes = plt.subplots(figsize = (18,18))
fn=['CompPrice', 'Income', 'Advertising', 'Population', 'Price',
       'ShelveLoc', 'Age', 'Education', 'Urban', 'US']
cn=['Low','Medium', 'High']

tree.plot_tree(model,
               feature_names = fn,
               class_names=cn,
               filled = True,
               fontsize=13);
```

```python
model.feature_importances_
```

```python
# predicting on test data set 
preds = model.predict(x_test)
```

```python
preds
```

```python
pandas.Series(preds).value_counts()
```

```python
pandas.crosstab(y_test,preds)
```

```python
# Accuracy 
numpy.mean(preds==y_test)*100
```

# Building Decision Tree Classifier (CART) using Gini Criteria


```python
model_gini=DecisionTreeClassifier(criterion="gini",max_depth=3)
```

```python
model_gini.fit(x_train,y_train)
```

```python
#PLot the decision tree
plt.figure(figsize=(18,10))

tree.plot_tree(model_gini);
```

```python
fig, axes = plt.subplots(figsize = (18,18))
fn=['CompPrice', 'Income', 'Advertising', 'Population', 'Price',
       'ShelveLoc', 'Age', 'Education', 'Urban', 'US']
cn=['Low','Medium', 'High']

tree.plot_tree(model_gini,
               feature_names = fn,
               class_names=cn,
               filled = True,
               fontsize=12);
```

```python
pred=model_gini.predict(x_test)
pandas.Series(preds).value_counts()
```

```python
pandas.crosstab(y_test,pred)
```
